### 2.2 SGD (Stochastic Gradient Descent)

```python
import numpy as np   # numeriacal computing
import matplotlib.pyplot as plt  # plotting core

w = np.linspace(-1,3,100)
# print(w)
# print(len(w))
b = 0

x = 2
y = 2

j = np.zeros(100)
# # print(j)
for i in range(len(w)):
    y_hat = w[i]*x + b;
    j[i] = 0.5 * (y_hat - y)**2

plt.plot(w,j, 'o' )
plt.plot(w,j, 'r-' )

plt.grid(True)
plt.show()
```

```python
import numpy as np   # numeriacal computing
import matplotlib.pyplot as plt  # plotting core

w = 1
b = np.linspace(-3,3,100)
x = 2
y = 2
j = np.zeros(100)
# print(j)
for i in range(len(b)):
    y_hat = w*x + b[i];
    j[i] = 0.5 * (y_hat - y)**2

plt.plot(b,j, 'o' )
plt.plot(b,j, 'r-' )

plt.grid(True)
plt.show()
```
