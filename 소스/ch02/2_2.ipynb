### 2.2 SGD (Stochastic Gradient Descent)

```python
import numpy as np   # numeriacal computing
import matplotlib.pyplot as plt  # plotting core

# simple function to demo step size
X=2
Y=2
b=0

def f(w) : # A parabola
    y_hat = w*X+b
    f = 0.5*(y_hat-Y)**2
    return f
def Df(w) : # The derivative (gradient)
    y_hat = w*X+b
    Df = X*(y_hat-Y)
    return Df
def xp1(w,alpha) : # update
    w = w - alpha * Df(w)
    return w

def plot_steps( guess, alpha, nsteps) :
    fig, ax = plt.subplots()
    x = np.linspace(-1,3,100)
    ax.plot(x, f(x))
    x = guess
    ax.plot(x,f(x), 'o', label='start x=%.2f' %x )
    for i in range(nsteps):
        xold = x
        x = xp1(x,alpha)
        #ax.plot(x,f(x), 'o', label='x = %.2f' %x)
        ax.plot(x,f(x), 'o')
        ax.plot([xold,x],[f(xold),f(x)], '-')
    plt.legend()
    plt.show()
    
plot_steps( 3, 0.03, 50)
```

```python
import numpy as np   # numeriacal computing
import matplotlib.pyplot as plt  # plotting core

# simple function to demo step size
def f(b) : # A parabola
    w=1
    x=2
    y=2
    y_hat = w*x+b
    f = 0.5*(y_hat-y)**2  
    return f
def Df(b) : # The derivative (gradient)
    w=1
    x=2
    y=2
    y_hat = w*x+b
    Df = y_hat-y  # (1*2+b-2) => (y_hat-y)*1
    return Df
def xp1(b,alpha) : # update
    b = b - alpha * Df(b)
    return b

def plot_steps( guess, alpha, nsteps) :
    fig, ax = plt.subplots()
    x = np.linspace(-3,3,100)
    ax.plot(x, f(x))
    x = guess
    ax.plot(x,f(x), 'o', label='start x=%.2f' %x )
    for i in range(nsteps):
        xold = x
        x = xp1(x,alpha)
        #ax.plot(x,f(x), 'o', label='x = %.2f' %x)
        ax.plot(x,f(x), 'o')
        ax.plot([xold,x],[f(xold),f(x)], '-')
    plt.legend()
    plt.show()
    
plot_steps( -3, 0.09, 50 )
```
